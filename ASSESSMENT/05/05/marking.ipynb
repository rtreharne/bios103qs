```python
import pandas as pd
```

```python
def read_file(fname):
    return pd.read_csv(fname)

def create_summary(data):
    new_data = data.copy()
    new_data["Location"] = "ALL"
    all_data = pd.concat([data, new_data])
    all_data["vbc"] = all_data["Viable counts"] * 2e5/1e7
    # Summary table with mean and standard deviation
    summary_table = all_data.groupby('Location')[['Temperature', 'Moisture', 'vbc']].agg(['mean', 'std'])

    # Apply rounding: 1 decimal place for Temperature and Moisture, 2 decimal places for vbc
    summary_table['Temperature'] = summary_table['Temperature'].round(1)
    summary_table['Moisture'] = summary_table['Moisture'].round(1)
    summary_table['vbc'] = summary_table['vbc'].round(2)

    return summary_table
```

```python
import os

# Get the current working directory
directory = os.getcwd()

# Get all files ending with .csv in the current directory
csv_files = [f for f in os.listdir(directory) if f.endswith('.csv')]
```

```python
csv_files = csv_files[101:202]
csv_files.append("compost_999.csv")
```

```python
import re
```

```python
tables = []

for fname in csv_files:
    data = read_file(fname)
    summary = create_summary(data)
    summary["question_id"] = fname
    tables.append(summary)
```

```python
full_summary = pd.concat(tables)
```
